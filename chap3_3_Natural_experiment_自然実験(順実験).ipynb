
# 自然実験(Natural experiment)
- 自然実験とは、RCTが実施されていない状況において、RCTが行われたかのような事象の発生を利用して因果効果を推定する方法
- 自然実験の分析手法は、それぞれ因果推論を成立させるための特定の仮定を前提としている
- それぞれの手法について、その仮定を以下に記載

---

### **1. 回帰不連続デザイン (RDD: Regression Discontinuity Design)**  
#### 仮定:
- **連続性の仮定 (Continuity Assumption)**:  
  処置変数（例えば、政策やプログラム）が適用される閾値の直前後で、処置以外の要因が滑らかに変化していること。  
  → つまり、処置の有無を除けば、閾値の直前後で対象者は同質であると仮定します。
- **正確な閾値の知識**:  
  閾値が明確であり、誤差なく観測可能であること。
- **操作可能性の排除 (No Manipulation)**:  
  閾値の直前後で対象者が意図的に自分のスコアを操作して処置を受ける/回避する行動を取らないこと。

#### 適用例:  
教育試験のスコアを基準に奨学金を提供する政策など。

---

### **2. 操作変数法 (IV: Instrumental Variables)**  
#### 仮定:
- **関連性の仮定 (Relevance)**:  
  操作変数が説明変数（処置変数）に影響を与えること。操作変数と説明変数の間には有意な関連性がある必要があります。
- **排除制約の仮定 (Exclusion Restriction)**:  
  操作変数が従属変数（結果変数）に影響を与えるのは、説明変数（処置変数）を通じてのみであること。操作変数が他の直接的な経路で結果変数に影響を与えないことを仮定します。
- **独立性の仮定 (Independence/Exogeneity)**:  
  操作変数は、未観測の交絡因子と独立であること（操作変数がランダムに割り当てられているか、交絡因子に依存しないこと）。

#### 適用例:  
兵役の抽選結果を操作変数として教育の因果効果を推定する研究など。

---

### **3. 差分の差分法 (DiD: Difference-in-Differences)**  
#### 仮定:
- **平行トレンド仮定 (Parallel Trends Assumption)**:  
  処置群と対照群が、処置が行われなかった場合でも、時間経過に伴うアウトカムのトレンドが同じであること。
  → この仮定が成立していれば、処置群の変化から対照群の変化を差し引くことで、処置の因果効果を推定可能です。
- **群の不変性**:  
  処置群と対照群が時間とともに変化しない（例えば、構成員が大幅に変わらない）こと。
- **外生的な介入**:  
  処置が無作為的または外生的に割り当てられていること。

#### 適用例:  
最低賃金政策が雇用率に与える影響を、政策導入前後の雇用率を比較して評価する場合など。

---

### **4. 合成コントロール法 (Synthetic Control Method)**  
#### 仮定:
- **加重線形独立性の仮定 (Weighted Linear Independence)**:  
  対象となる処置群と対照群の特徴が、対照群の加重線形結合で再現可能であること。
- **外挿性の仮定 (Extrapolation Assumption)**:  
  処置群が処置を受けなかった場合の反実仮想を、対照群の加重によって近似できること。
- **外生的介入の仮定**:  
  処置が外生的に割り当てられており、他の変化と独立であること。

#### 適用例:  
政策やプログラムの導入が単一地域や一部のグループに限られている場合（例えば、特定州のタバコ税導入の効果など）。

---

### **要点まとめ:**
| 分析手法 | 主な仮定 |
|---|---|
| RDD | 連続性の仮定、操作可能性の排除 |
| IV | 関連性、排除制約、独立性 |
| DiD | 平行トレンド、外生的介入 |
| 合成コントロール法 | 加重線形独立性、外挿性、外生的介入 |
